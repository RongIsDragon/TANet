```python
# 导入库函数
from pycocotools.coco import COCO
import numpy as np
import skimage.io as io
import matplotlib.pyplot as plt
import pylab
import cv2
pylab.rcParams['figure.figsize'] = (8.0, 10.0)
```

```python
# 读取标注数据
annFile = '/home/omnisky/_disk/chengming/modanet-master/annotations/modanet2018_instances_val.json'
coco = COCO(annFile)
```

```python
imgIds = coco.getImgIds()
for imgId in imgIds:
    annIds = coco.getAnnIds(imgIds=imgId, iscrowd=None) 
    anns = coco.loadAnns(annIds) 
    print(anns)
    masks = [] 
    showonce = True 
    if anns == []:    # 图片无标注数据
        continue
    for ann in anns:
        m = coco.annToMask(ann)
        m = m * ann['category_id']
        catids.append(ann['category_id'])
        masks.append(m)
    if masks[0].shape != (600, 400):    # 图片标注不为600 * 400
        continue
    seg = np.zeros(masks[0].shape)
    for mask in masks:
        seg += mask
    seg[~(np.isin(seg, catids))] = 0
    if np.max(seg)>13 or np.min(seg)<0:
        print('no')
    cv2.imwrite('../data/val/labels/' + str(imgId) + '.png', seg)
#     print(str(imgId))
#     plt.imshow(seg)
#     plt.show()
```
